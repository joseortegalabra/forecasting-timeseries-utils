```python
# a la última fecha restar 28 días
day_last_date = int(data_last_date.split('_')[1])
day_start_date = day_last_date - 28
day_start_date = day_start_date + 1 # se necesita sumar un día para tomar los últimos 28 días incluyendo el último
data_start_date = 'd_' + str(day_start_date)
data_start_date
```
